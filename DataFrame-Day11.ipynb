```python
df1
```

```python
df2
```

```python
df_combine = pd.merge(df1,df2,on="branch",how="right")
df_combine
```

```python
df_combine = pd.merge(df1,df2,on="branch",how="left")
df_combine
```

```python
df_combine = pd.merge(df1,df2,on="branch",how="right",indicator=True)
df_combine
```
